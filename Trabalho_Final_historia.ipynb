  **Lore de Drakulor:**

  Nos confins do mundo, onde o oceano encontra as montanhas e o sol se põe em um espetáculo de cores ardentes, jaz a Baía dos Dragões. Este é um lugar sagrado, onde os dragões antigos, ao fim de suas vidas, voavam para descansar e deixar seus corpos mortais se fundirem com a terra e o mar.

  Por séculos, a Baía dos Dragões permaneceu intocada, um santuário onde a passagem dos grandes seres alados para o além era honrada e respeitada. No entanto, o destino sombrio começou a se desenrolar quando um mago louco, consumido pela sede de poder e conhecimento proibido, descobriu a localização secreta da Baía.

  Ignorando os avisos dos sábios e das criaturas antigas que guardavam a Baía, o mago profanou o local sagrado em sua busca por segredos ocultos e tesouros esquecidos. Ele perturbou os restos mortais dos dragões, despertando o ódio e a ira dos espíritos ancestrais que habitavam ali.
    
  </p>
  </div>
</div>

<div class="texto-abaixo">
  <p>

  Feridos pela profanação de seu descanso eterno, os dragões mortos não encontraram a paz na morte. Em vez disso, eles se ergueram como criaturas de ossos conduzidos por uma fúria ancestral. Drakulor, o Dragão de Ossos, foi um desses seres que emergiram da Baía com um rugido de fúria e vingança.

  Seus ossos eram praticamente indestrutíveis, impregnados com a magia antiga que os dragões acumularam ao longo de suas vidas imortais. Drakulor era uma força da natureza, uma tempestade de ossos e ódio que varria tudo em seu caminho.

  Com um rugido que ecoava pelos céus e os mares, Drakulor espalhou o terror e a destruição por onde passava. Ele não conhecia a piedade, nem a compaixão; sua única missão era vingar-se da profanação de seu sagrado local de descanso e punir aqueles que ousaram desafiar os dragões.

  Apesar de sua natureza monstruosa, há uma tragédia na história de Drakulor, uma lembrança do respeito e da honra que um dia foram mostrados aos dragões antigos. Agora, ele é uma lembrança viva do perigo de perturbar os mortos e desafiar os antigos poderes que governam o mundo.

  </p>
</div>



<div class="container">
  <div class="texto-right">
  <p>

  **Durin, o Martelo-de-Ferro:**

  O anão, chamado Durin Martelo-de-Ferro, era conhecido em todo o reino como um habilidoso ferreiro e um amigo leal. Quando soube da transformação de Drakulor em um monstro de ossos, Durin sentiu um aperto no coração, lembrando-se dos dias de glória e camaradagem que compartilhara com o antigo dragão.

  Determinado a ajudar seu velho amigo a encontrar a paz, Durin partiu em uma jornada em busca de um material tão poderoso quanto os ossos de dragão, para forjar uma espada capaz de dar fim ao lamento de Drakulor.

  Durin viajou por terras distantes, explorando cavernas escuras e montanhas escarpadas em busca do material ideal. Ele enfrentou perigos inimagináveis e desafios intransponíveis, mas sua determinação era tão firme quanto o aço que manipulava em sua forja.
    
  </p>
  </div>
  <img class="imagem" src="Durin.jpeg" alt="imagem monstro">
</div>

<div class="texto-abaixo">
  <p>

  Finalmente, após meses de busca incansável, Durin encontrou o que procurava nas profundezas de uma montanha vulcânica: uma liga de metal negro, temperada pelo calor das entranhas da terra e imbuida com a energia elemental do fogo, capaz de causar um dano extremo a espiritos malignos, e da terra, capaz de resistir a maldições.

  Este metal é conhecido como Ébano Infernal, mas Durin não sabia se ele era tão duro quanto os ossos de dragão. Com ele, o ferreiro anão forjou uma espada que brilhava com uma luz sombria e pulsante, mas ele queria saber qual a dureza de sua espada.

  Assim Durin, com a Espada do Ébano Infernal em mãos, foi até a torre magica do continente leste, onde o reino magico de CENEPEM tem a maior influencia, procurando algum feitiço de pudesse ajuda-lo. 

  </p>
</div>

  <p>

  **Torre Magica de CENEPEM:**

  A Torre Arcana de CENEPEM erguia-se majestosamente no coração do reino mágico, uma estrutura imponente que alcançava os céus com suas torres espirais e cúpulas reluzentes. Construída há séculos pelos mais habilidosos magos da época, a torre era um símbolo de sabedoria, poder e magia no continente leste.

  No interior da torre, corredores misteriosos se entrelaçavam em uma intricada teia de salas de estudo, laboratórios alquímicos e salões de cerimônia. Cada andar da torre abrigava segredos arcanos e conhecimentos ancestrais, preservados e protegidos pelos magos que dedicavam suas vidas ao estudo das artes mágicas.

  No topo da torre, sob um céu estrelado e pulsante, ficava o Sanctum Arcanum, o santuário dos mais poderosos magos do reino. Lá, o Trio Arcano - Sirius, Bhioranovateus e Nantemius - reunia-se para conduzir rituais poderosos, consultar as estrelas em busca de orientação e defender o reino contra as forças das trevas.
    
  </p>
  </div>
</div>

<div class="texto-abaixo">
  <p>

  Ao redor da torre, jardins encantados e fontes de água cristalina acrescentavam uma aura de magia e encanto ao ambiente, enquanto sentinelas mágicas mantinham vigilância constante contra qualquer ameaça que se atrevesse a se aproximar.

  A Torre Arcana de CENEPEM era mais do que um simples edifício; era o coração pulsante da magia no continente leste, um farol de conhecimento e poder que guiava os estudiosos e os aspirantes a magos em sua jornada em busca da iluminação e do domínio das artes arcanas. E através dos séculos, seu legado perduraria, inspirando e protegendo aqueles que buscavam a luz da sabedoria mística.

  Ao chegar à torre mágica do continente leste, Durin foi recebido pelos três magos mais renomados do reino de CENEPEM: Sirius, o Mestre das Estrelas; Bhioranovateus, a Sábia dos recursos; e Nanomius, a Guardiã dos Elementos. Juntos, eles eram conhecidos como o Trio Arcano, e suas habilidades mágicas eram incomparáveis em todo o continente.

  Durin explicou sua missão aos magos, revelando o motivo de sua busca pelo feitiço que poderia testar a dureza da Espada do Ébano Infernal. Os magos ouviram atentamente, compreendendo a importância da tarefa que o anão tinha diante de si.

  Após uma breve deliberação, os magos concordaram em ajudar Durin. Sirius utilizaria seus conhecimentos estelares para testar a energia cósmica das estrelas na espada. Bhioranovateus empregaria suas conhecimentos a respeito dos recursos para comparar o Ébano com outros materiais. Enquanto isso, Nanomius invocaria os elementos da terra e do fogo para analisar o Ébano Infernal.

  Juntos, os três magos conjuraram um poderoso feitiço, envolvendo a espada de Durin em um brilho místico e irradiante. Quando a magia se dissipou, a lâmina da espada brilhava com uma intensidade incomum, emitindo uma aura de poder e proteção.

  Com um sorriso de gratidão, Durin agradeceu aos magos por sua ajuda, sabendo que agora estava preparado para enfrentar Drakulor com a Espada do Ébano Infernal em mãos. Com a confiança renovada e determinação inabalável, ele partiu em sua jornada para a Baía dos Dragões, onde esperava encontrar seu velho amigo e libertá-lo do seu sofrimento eterno.

  </p>
